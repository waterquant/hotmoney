## “减持”杀手降临龙头股，最应该做的事是不要幻想

*  减持，已经成为市场人气最大的杀手，不要对即将减持的股票抱有任何幻想。曾经有过幻想的小伙伴全部都在恒立实业中阵亡了。所以，明天将会是近期亏损效应最大的一天，所有的强势股都将会拉起警报。**重点要做好获利止盈与防御，而不是老想着进攻。**

* 光洋股份，这么牛逼的票，为啥一直没有加入我的自选股呢？**因为这坑爹的股权分布啊**。股权比例极其分散，5%以上的股东一共有5个。这简直就是减持的高危品种。而看一下市北高新，就知道什么叫做良心的股权结构了。市北高新集团绝对控股。且两年以来只有增持，没有减持。所以，很多朋友说，市北高新可以当做新股炒，就是这个原因。因为除了实际控制人别的持股都非常少，股权结构就类似于刚上市的新股。

## 明日局面推演

* 光洋股份已经确定凉凉，就算不是跌停开，也基本跌停或者全天深跌。那么这意味着，市场中最后一个龙头像的公司遭遇重创。我们回顾一下本次短线行情的最强的股票情况。恒立实业自不必说，被减持已经彻底打垮，天洋股份也即将被减持打败。亚振家具，西安旅游，民丰特纸，群兴玩具全部覆没。


  目前， 除了市北高新之外，全部都挂了。而市北之所以这么强，真的是因为筹码结构好，但是，明天呢？
  
  
  **已经横盘了两日，资金实际上立刻就要做出决断，但是目前来看，假设你持有市北。如果明天光洋股份跌停了，而整个创投板块全部都跟着下跌，那你会不会有卖出的冲动？如果你持有光洋股份跌停卖出了，你是否还有勇气买市北？如果你今天空仓了，你看到光洋跌停了，你是否会开仓市北？**
  
  
  问完这三个问题，我想大家都知道了。明天只要有大资金开始砸盘，那么最终的结果一定非常惨烈。当然，还有一个希望。就是欢乐海岸这样级别的游资，强行买入，让所有资金都集中精力来做市北。这是唯一有可能有利的剧本，但不取决于我们，取决于大柚子的决心。
