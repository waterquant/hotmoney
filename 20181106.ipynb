* 为什么呢？因为你就是把牛皮吹爆，目前所有的概念股都只是概念。都不可能会迅速的产生业绩。


* 是的，目前市场最牛逼的概念股都只是概念股。大家一定一定要把这句话记在心里。一旦市场的情绪出现反转，千万不要自欺欺人。许多人曾经因为千年大计的幻想，而深陷泥潭。


* 把故事当做现实，最后现实会狠狠地抽你的巴掌。好了，预防针打好了，咱们在这个前提下，做出一些对比。

    首先，本次创投行情和雄安行情具有非常高的相似度：

     第一，都是突然宣布，事先广大小散全部都不知情。

     第二，虽然突然宣布，但很明显已经有权贵资本进行埋伏。

     第三，行情初期，最核心的概念股全部都是一字板。

     看到这里，是不是觉得相似度非常高。但是，也有很不同的地方，那就是，雄安板块其实是没有后续的，但创投行情来自于科创板，因此是有后续的。而且科创板行情本身直接作用于资本市场。

     因此，我们基本可以判断：

     这波创投行情虽然开始的也很迅猛，但是一开始的强度一定不如雄安。但是，一定会比雄安更持久，且一定会反复做很多波，一直到科创板的征求意见稿正式落地。而落地之后，将会进入另一个行情周期。

     换言之，科创板行情将持续一个大周期，而创投行情只是科创板行情的一个开始。甚至不一定是高度最高的行情。但毫无疑问，第一波出货的时候，都将是非常惨烈的。
