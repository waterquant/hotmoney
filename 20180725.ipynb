### 涨停，A股的顶级逻辑

* 神奇的A股最神奇的制度就是涨跌停板制度

* 涨跌停板制度最初的目的是防止股票过度投机，造成大起大落。这听起来非
常美。但是这个非常美好初衷的的制度，造成了一个非常重要的问题：

* 股票的流动性问题

* 无论是跌停板，还是涨停板，都会造成一个问题。就是流动性的枯竭，从而
想卖卖不出，想买买不到。
当你想买一只股票，买不到的时候会怎么办？会造成哄抢，从而造成更多的
恐慌，引发更多的上涨冲力。
当你想卖一只股票，卖不掉的时候怎么办？会造成更多的恐慌性抛售，从而
引发更大的下跌冲击力。
其实A股的历次股灾都是流动性危机，而流动性危机的根源就在涨跌停制度。
2016 年底实行的熔断制度，指数级别的涨跌停，事实上就只能造成更大的危机。
