## 市北高新的40亿是怎么算出来的？

*  一只股票的成交额有多重要？是获利的生命之源。计算一只股票的容量非常重要。



   因为，我们都知道，股票之所以会涨停，是因为买的力量远大于卖的力量。而一只股票的理论最大成交额直接代表着这只票的承载力量。



    这就是，次新股哪怕70%的换手，第二天还可能继续涨停。但如果是老股票，放巨量的时候也意味着坑人的时候。

  

   我知道，今天很多朋友都会问我，你是怎么估算出来，市北高新的巅峰成交额是四十亿的?



   现在，我把方法交给大家。主要考虑因素如下：



   **1，股票高度。股票高度越高的话，前期所有的套牢资金都可能卖出，因此，在越高的位置开板则换手越大。**



   **2, 80%定律。市场极限换手的上限为80%。**



   **3，实际换手情况。比如有些公司虽然流通值很大，但如果这些流通值都是超过5%以上的股东持有，那么实际上是非流通的。**
   
    比如市北高新的第一大股东拿了7.71亿的流通股，这些流通股实际上是不流通的。考虑到市北高新虽然连续十一板，但是巅峰时候的价格是22元，且目前的价格只是17年年初的价格。因此，虽然创新高，但并不能达到80%的上限。



   所以我折合了一下，给了80%的比率。也就是20%的套牢盘不愿意割肉。所以就是 **实际流通值 乘以 80% 再乘以 80%愿意卖出的比率的**，算下来大概就是四十亿的成交额。

    而今天离四十亿还差一点点。原因可能是，还未触发到之前套牢盘的位置，以及少部分资金认为还有继续涨停的可能性，打算明天再卖。虽然这部分资金很少，但都会转化成为明天的抛盘。

    总而言之，你问我明天市北会怎么走。我只能说，自从2018年以来，所有的妖股的极限承载量就是20亿，超过了20多个亿，全部都死掉了，且都没有第二波。所以，你如果继续看好，那么你就要给自己一个理由。给出一个例外的理由。而且，市北高新还是连续一字板之后的37.6亿的顶部巨量。同时，还卖出了自己作为炒作热点的核心资产。
    
    当然，我在这里说再多又没什么用。因为毕竟有近40亿的资金接盘。我觉得大家可以想一想，自己明天愿不愿意为这几十亿的资金接盘，这才是最重要的。

    因为短线本身就是零和游戏。你自己的股票总有人来接盘，你才能获利。所以，想一想，明天你愿意来市北吗？

## 为什么笃定恒立实业不是欢乐海岸

*  我觉得今天最欣慰的事情，应该就是恒立实业的这一波非常到位的提示。说一句到位，真的不过分。


     用脑子想一想，就知道肯定不是欢乐海岸。我觉得应该绝大多数朋友都应该明白。


     至于少数看了所谓欢乐海岸订阅号的朋友，今天一头扎进去，我只能说什么好呢。看在今天亏钱的份儿上，我就不嘲讽了。吃一堑长一智。用脚想都知道几十亿资金规模的欢乐海岸专门开订阅号教小散炒股？生怕证监会查不到他们操纵股价的证据是吧。
     
     
     不过，这也说明一个问题，市场的韭菜还是非常多的。所以，这里祝所有的“欢乐海岸”+“赵老哥”+“炒股养家”们的订阅号，阅读量越高越好，点赞数越高越好，早点十万加。让市场的韭菜更多一点。


     当然，前提是，我希望我所有的读者都是镰刀而不是韭菜。


  那么回到一个问题，为什么昨天的恒立实业肯定不是欢乐海岸。
     
     
 原因很简单， **每一种成名游资，他们的交易模式都是非常固定的。越是大游资，越是固定**
     

  这就好比是练武功一样，**武功越高的人，武功的路数越纯**。因为不同路数的武功越练到高深的时候，区别越大。
     
      
 **所以，专注于龙头股封板模式的欢乐海岸，是觉得不会去低吸一只全天都在水下的曾经的龙头股**
     
     
  尤其是在寻找到真龙之后。欢乐海岸既然已经选了鲁信创投，就绝不会低吸恒立实业。就算要买恒立实业也会标志性封涨停。


  偷偷摸搞低吸，然后第二天拉升就大笔出货。这是二线游资干的事情，绝不是龙头股的缔造者欢乐海岸干的事情。

## 为什么天风证券尾盘会砸开

*  今天最正确的事情，应该就是忍住了天风证券没有买。其实午盘的时候，我已经说的很清楚了。


   想上，但是没有上。说到底，还是对于风险的防御。看今天的龙虎榜，你会特别感慨。


   居然是被小资金给砸崩的。是的，但是我告诉大家这更可怕。



   因为这代表着人气的下滑。得散户者得天下，但是散户已经撤在了游资前面了。为什么会撤？



   第一，对于指数的担心。整个券商都在回调。


   第二，天风最大的亮点在于创投。对于创投的担心。



   为什么会担心？因为今天市北巨量，明天会怎么样？
   


    当然，证券股其实并不算是结束，只是今天的压力确实特别大。但还没有放量。
    
    

    但是我今天还是宣布我空仓了，连国海证券都没有剩下，是因为我不太想再接受调整了。券商的未来当然还可能继续延续趋势，但是我已经不想在接受可能的亏损，先把盈利保住。
    

     毕竟今年年底赚一点钱真的不容易。今天很多朋友感慨，上周赚的钱，这两天全部交出去了。保住利润真的很重要。

*  五、下一波创投怎么做？



   简单来说，下一波创投，要炒就要炒实锤了。



   我的预感是，下一波的龙头，不在目前所有的已经高位的股票中。包括现在已经很高的市北、鲁创。而是一只不起眼且全新的票。



    因为实锤已经越来越近了。



    我们接下来，就是找出可能的50家，以及参股的50家。也就是从现在的创投变成更加具体的科创板个股挖掘。



    所以，我觉得那些蹭热点的也全部都应该放弃了。这一个大新闻，才是接下来可能的主流。至于具体哪些股票，有待市场发掘，我明天早知道也会公布一些我自己找的个股，但是仅供参考。



    六、总结



    1，前期所有的创投龙头暂时都全部冻结，我暂时看不到任何一只有潜力继续带领市场。



     2，第二波很快就会开启，本周内一定有所动作，也许明天就会有个股异军突起。且我不认为是老股票。



     3，明天依然是腥风血雨，亏钱效应会继续放大，但也存在一些股票的反包。第二波的带动也许就是从反包开始的。



     4，第二波行情不会是普涨，只是少数股票的表演，所以不要指望第二波可以把第一波的你已经套牢的个股带动起来。找到好位置，赶紧走人是最好的选择。至于雄安，我从来都没有这个选项。
